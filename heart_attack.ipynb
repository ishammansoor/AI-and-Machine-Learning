```python
#importing the needed libraries
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras import layers
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

Reading in the data file using pandas, then spliting it into traing and testing set to later train the data

```python
df = pd.read_csv("/content/Medicaldataset.csv")


training_set = df[:int(len(df) * .8)]
testing_set = df[int(len(df) *.8):]

testingY = testing_set.pop("Result")


X = training_set.drop(columns=["Result"])
y = training_set["Result"]

y = y.map({
    'negative': 0,
    'positive': 1,
})

#spliting into validation and traning sets
trainX, valX, trainY, valY = train_test_split(X, y, test_size=0.25, random_state=42)

input_shape = [X.shape[1]]

X = np.array(X)
y = np.array(y)

testing_set = np.array(testing_set)
```

```python
model = keras.Sequential([
    layers.BatchNormalization(input_shape=input_shape),
    layers.Dense(456, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(rate=0.3),
    layers.Dense(456, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(rate=0.3),
    layers.Dense(456, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(rate=0.3),
    layers.Dense(456, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(rate=0.3),
    layers.Dense(356, activation="relu"),
    layers.BatchNormalization(),
    layers.Dropout(rate=0.3),
    layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["binary_accuracy"]
)

model.fit(
    trainX, trainY,
    validation_data=(valX, valY),
    batch_size=512,
    epochs=650,
)
```

Testing the model

```python
prediction = model.predict(testing_set)

mapped = (prediction > 0.5).astype('int32')

for i, j in zip(prediction, testingY):
  print("Expected", j, "Predictect: ", i)
```
